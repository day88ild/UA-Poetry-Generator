```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import tensorflow as tf
import warnings
warnings.filterwarnings("ignore")
```

```python
df_taras = pd.read_csv("data/taras_shevchenko_poetry.csv", index_col=0)
df_taras.head(30)
```

```python
df_taras.info()
```

```python
# Create a heatmap to visualize missing values in the DataFrame 'df_taras'
plot = sns.heatmap(df_taras.isnull())

# Set the size of the figure (plot)
plot.figure.set_size_inches(15, 10)
```

```python
# Drop rows with missing values from the DataFrame 'df_taras'
df_taras.dropna(inplace=True)
```

```python
# show the first poetry from the DataFrame 'df_taras'
print(df_taras.poetry_text[0])
```

```python
# Import the 'detect' function from the 'langdetect' library
# detect function is used to find the language of text
from langdetect import detect
```

```python
# create new column 'language' in DataFrame 'df_taras' that contains information about language of the written poetry
df_taras["language"] = df_taras.apply(lambda row: detect(str(row.poetry_text)), axis=1)
```

```python
# Calculate the counts of each language detected in the DataFrame column 'language'
df_taras.language.value_counts()
```

```python
# create list of 'uk' poetry 
texts = list(df_taras[df_taras.language == "uk"].poetry_text)
```

```python
print(texts)
```

```python
# Create a Tokenizer object for text preprocessing
tokenizer = tf.keras.preprocessing.text.Tokenizer(oov_token="<OOV>", filters='‘1234567890’»«“–—”…!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n')
tokenizer.fit_on_texts(texts)
```

```python
word_index = tokenizer.word_index
```

```python
# print the number of unique words in our text
num_unique_words = len(word_index)
num_unique_words
```

```python
# create sequences from texts and print the first example
sequences = tokenizer.texts_to_sequences(texts)
sequences[0]
```

```python
# Initialize an empty list to store input sequences
input_sequences = []
```

```python
# Iterate through each sequence in the 'sequences' list
for sequence in sequences:
    for i in range(2, len(sequence)):
        # Append a subsequence of the current sequence to the 'input_sequences' list
        input_sequences.append(sequence[:i])
```

```python
# Print the value of the third element
input_sequences[2]
```

```python
# print the maximum lenth of subsequences
max(map(len, input_sequences))
```

```python
# Padding the input sequences to have a maximum length of 100 tokens
input_sequences = tf.keras.preprocessing.sequence.pad_sequences(input_sequences, maxlen=100)
```

```python
# Print the value of the third element again
input_sequences[2]
```

```python
# print number of subsequences
len(input_sequences)
```

```python
# shuffle the list of subsequences
np.random.shuffle(input_sequences)
```

```python
# creating list of labels and sequences
input_labels = input_sequences[:, -1]
input_sequences = input_sequences[:, :-1]
```

# function for generating poetry

```python
# creating decoder for sequences
word_index_rev = {value: key for key, value in word_index.items()}
```

```python
# create function to generate poetry using model
def gen_poetry(first_words, model, num_word):
    # Convert the input words to a sequence of token IDs using the tokenizer
    sequence = tokenizer.texts_to_sequences([first_words])
    # creating the list of texts that will be output text
    text = first_words.split()
    # padding sequences to have maximum lenth of 99 tokens
    sequence = tf.keras.preprocessing.sequence.pad_sequences(sequence, maxlen=99)
    # iterate 'num_word' times
    for _ in range(num_word):
        # predict next word
        last_word = model.predict(sequence[:, -99:])
        # adding last word to 'sequence' and to 'text'
        sequence = np.append(sequence, last_word.argmax())[None, :]
        text.append(word_index_rev[last_word.argmax()])

    return " ".join(text) 
```

# model training

```python
# creating model
model_taras_sh_scratch = tf.keras.Sequential([tf.keras.layers.Embedding(input_dim=num_unique_words+1, output_dim=32, input_length=99),
                                              tf.keras.layers.Bidirectional(tf.keras.layers.GRU(16)),
                                              tf.keras.layers.Dense(units=256, activation="relu"),
                                              tf.keras.layers.Dense(units=num_unique_words+1, activation="softmax")])
# compile the model
model_taras_sh_scratch.compile(optimizer=tf.keras.optimizers.Adam(),
                               loss="sparse_categorical_crossentropy",
                               metrics="acc")
```

```python
# load model if it exists
model_taras_sh_scratch = tf.keras.models.load_model("models/model_taras_sh_scratch.h5")
model_taras_sh_scratch
```

```python
# printing the summaty of the model
model_taras_sh_scratch.summary()
```

```python
# training
history = model_taras_sh_scratch.fit(input_sequences, input_labels, validation_split=0.2, epochs=100, batch_size=1024)
```

```python
# saving the model
model_taras_sh_scratch.save("models/model_taras_sh_scratch.h5")
```

```python
# plotting the history of training
fig, axis = plt.subplots(1, 2, figsize=(15, 4))

axis[0].plot(history.history["loss"], label="train")
#axis[0].plot(history.history["val_loss"], label="val")
axis[0].set_title("loss - val_loss")
axis[0].set_xlabel("epoch", fontsize=13)
axis[0].legend(bbox_to_anchor=(1.0, 1.0))


axis[1].plot(history.history["acc"], label="train")
#axis[1].plot(history.history["val_acc"], label="val")
axis[1].set_title("acc - val_acc")
axis[1].set_xlabel("epoch", fontsize=13)
axis[1].legend(bbox_to_anchor=(1.0, 1.0))


plt.show()
```

```python
# first words of our future poetry
gen_text = """У нашім раї на землі"""
# generating the poetry
text = gen_poetry(gen_text, model_taras_sh_scratch, 100)
```
